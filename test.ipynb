```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib as mpl
# Increase the chunksize
mpl.rcParams['agg.path.chunksize'] = 1000  # Set this to a higher value if needed
```

```python
df = pd.read_csv('Input_Data/vmCloud_data.csv', parse_dates=['timestamp'])
```

```python
print(df.size)
```

```python
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Handle missing values in cpu_usage (forward fill for demonstration)
df['cpu_usage'] = df['cpu_usage'].fillna(method='ffill')

# Convert cpu_usage to float to ensure proper plotting
df['cpu_usage'] = df['cpu_usage'].astype(float)

# Plot the cpu_usage column against timestamp
plt.figure(figsize=(10, 5))
plt.plot(df['timestamp'], df['cpu_usage'], marker='o', label='CPU Usage', color='blue')
plt.title('CPU Usage Over Time')
plt.xlabel('Timestamp')
plt.ylabel('CPU Usage (%)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```
